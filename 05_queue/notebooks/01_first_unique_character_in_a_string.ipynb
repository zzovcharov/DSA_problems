## First Unique Character in a String

Given a string `s`, find the **first non-repeating character** in it and return its **index**.
If no such character exists, return `-1`.



### Examples

#### Example 1
**Input:**
s = "leetcode"
**Output:**
0

#### Example 2
**Input:**
s = "loveleetcode"
**Output:**
2

#### Example 3
**Input:**
s = "aabb"
**Output:**
-1



### Constraints
- `1 <= s.length <= 10^5`
- `s` consists of only lowercase English letters (`a`–`z`)





### Approach

1. Loop through the string from left to right using an index.
2. For each character `s[i]`, count how many times it appears in the string.
3. If a character appears **exactly once**, it is the first non-repeating character.
4. Return its index immediately.
5. If no such character is found after the loop, return `-1`.

### Complexity Analysis

Time Complexity: O(n²)

- Counting characters takes O(n) and is done for each character.

- Space Complexity: O(1)

No extra data structures are used.

```python
class Solution(object):
    def firstUniqChar(self, s):
        """
        :type s: str
        :rtype: int
        """
        for i in range(len(s)):
            if s.count(s[i]) == 1:
                return i
        return -1
```
