### Problem Description

The **k-beauty** of an integer `num` is defined as the number of substrings of `num` (when read as a string) that satisfy the following conditions:

1. The substring has a length of `k`.
2. The substring is a divisor of `num`.

**Notes:**

- Leading zeros are allowed.
- 0 is not a divisor of any value.
- A substring is a contiguous sequence of characters in a string.



### Examples

**Example 1:**

Input: num = 240, k = 2
Output: 2
Explanation:
Substrings of length 2: "24", "40"

- "24" divides 240 → yes

- "40" divides 240 → yes
k-beauty = 2




### Constraints

- 1 <= num <= 10^9
- 1 <= k <= length of num (when taken as a string)


### Approach

We want to count the number of substrings of length `k` in `num` that divide `num` evenly.

1. Convert the number `num` to a string `s`.
2. Initialize a counter `count = 0`.
3. Loop through all possible substrings of length `k` in `s`:
   - Extract the substring `sub = s[i:i+k]`.
   - Convert it back to an integer `val = int(sub)`.
   - If `val != 0` and `num % val == 0`, increment the counter.
4. Return `count` as the k-beauty of `num`.

```python
class Solution(object):
    def divisorSubstrings(self, num, k):
        """
        :type num: int
        :type k: int
        :rtype: int
        """
        s = str(num)
        count = 0

        for i in range(len(s) - k + 1):
            sub = s[i:i + k]
            val = int(sub)
            if val != 0 and num % val == 0:
                count += 1

        return count
```
