### Problem

A string is **good** if there are no repeated characters.

Given a string `s`, return the number of **good substrings of length three** in `s`.

- Every occurrence of a substring counts, even if it's repeated elsewhere.
- A substring is a contiguous sequence of characters in a string.



### Example 1

**Input:** `s = "xyzzaz"`
**Output:** `1`
**Explanation:** Substrings of size 3 are `"xyz"`, `"yzz"`, `"zza"`, and `"zaz"`.
Only `"xyz"` has all unique characters.

### Example 2

**Input:** `s = "aababcabc"`
**Output:** `4`
**Explanation:** Substrings of size 3 are `"aab"`, `"aba"`, `"bab"`, `"abc"`, `"bca"`, `"cab"`, `"abc"`.
Good substrings: `"abc"`, `"bca"`, `"cab"`, `"abc"`.



### Constraints

- `1 <= s.length <= 100`
- `s` consists of lowercase English letters.


### Approach

We want to count all substrings of length 3 that have **no repeated characters**.  

1. Initialize a counter `count = 0`.
2. Loop through the string `s` from index `0` to `len(s) - 3` (inclusive) to get all substrings of length 3.
3. For each substring:
   - Convert it into a set. A set only keeps **unique characters**.
   - If the length of the set is 3, all characters are unique â†’ increment the counter.
4. Return the final `count`.

```python
class Solution(object):
    def countGoodSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        count = 0
        for i in range(len(s)- 2):
            substring = s[i:i+3]
            if len(set(substring)) == 3:
                count +=1
        return count
```
