## Problem: Reverse Words in a String

Given a string `s`, reverse the order of characters in each word within a sentence while preserving whitespace and the original word order.

**Examples**

Input: "Let's take LeetCode contest"
Output: "s'teL ekat edoCteeL tsetnoc"

Input: "Mr Ding"
Output: "rM gniD"

**Constraints**

1 <= s.length <= 5 * 10^4
`s` contains printable ASCII characters.
`s` has no leading or trailing spaces.
There is at least one word in `s`.
All words are separated by a single space

## Approach

1. Split the string into words using spaces.
2. Reverse each word individually.
3. Store the reversed words in a list.
4. Join the reversed words back together with spaces.
5. Return the final transformed string.

Time complexity: O(n)
Space complexity: O(n)

```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        words = s.split(" ")

        reversed_words = []
        for word in words:
            reversed_words.append(word[::-1])

        return " ".join(reversed_words)
```

## Rubber Duck Explanation

We start with a normal sentence where the words are in the correct order, but each individual word needs its characters reversed. So we first split the sentence into separate words using the spaces as separators.

Now we take one word at a time and flip it backwards using slicing (`[::-1]`). Each reversed word gets added to a new list.

Once all the words are reversed, we put them back together with spaces in between â€” exactly like the original sentence layout. The word order stays the same, but each word has its letters reversed.
