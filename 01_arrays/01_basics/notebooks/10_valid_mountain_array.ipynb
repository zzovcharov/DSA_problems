## Problem: Valid Mountain Array

Given an array of integers arr, return true if and only if it is a valid mountain array.

An array is a mountain array if:

arr.length >= 3

There exists an index i (the peak) such that:

The sequence is strictly increasing from arr[0] to arr[i]

The sequence is strictly decreasing from arr[i] to arr[n-1]

Examples

**Example 1**
Input: arr = [2,1]
Output: false

**Example 2**
Input: arr = [3,5,5]
Output: false

**Example 3**
Input: arr = [0,3,2,1]
Output: true

## Approach

To check whether an array forms a valid mountain:

1. Check length

A mountain must have:

At least one element going up

A peak

At least one element going down

So if len(arr) < 3, return False.

```python
class Solution(object):
    def validMountainArray(self, arr):
        """
        :type arr: List[int]
        :rtype: bool
        """

        n = len(arr)
        if n < 3:
            return False
        i = 0
        #up
        while i + 1 < n and arr[i] < arr[i + 1]:
            i += 1
        #peak not first nor last
        if i == 0 or i == n - 1:
            return False
        # down
        while i + 1 < n and arr[i] > arr[i + 1]:
            i += 1
        return i == n - 1
```
