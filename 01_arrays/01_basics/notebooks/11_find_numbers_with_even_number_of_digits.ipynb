## Problem: Find Numbers with Even Number of Digits

You are given an array `nums` of integers.
Return **how many numbers contain an even number of digits**.



### Examples

**Example 1**

Input:
`nums = [12,345,2,6,7896]`
Output:
`2`

Explanation:
- 12 → 2 digits (even)
- 345 → 3 digits (odd)
- 2 → 1 digit (odd)
- 6 → 1 digit (odd)
- 7896 → 4 digits (even)

Only **12** and **7896** have an even number of digits.



**Example 2**

Input:
`nums = [555,901,482,1771]`
Output:
`1`

Explanation:
- Only **1771** contains an even number of digits.



### Constraints
- `1 <= nums.length <= 500`
- `1 <= nums[i] <= 10^5`





## Approach

To determine how many numbers contain an even number of digits, follow these steps:

Iterate through each number in the array.

Convert the number to a string so we can easily measure how many digits it has.

Count the digits using len(str(num)).

Check whether the digit count is even using count % 2 == 0.

If it is even, increase the total count.

Return the final count.

## Key Insight

A number has an even number of digits if:

len(str(num)) % 2 == 0

This keeps the solution simple and efficient.

### Time & Space Complexity

Time Complexity: O(n) — we inspect each number once.

Space Complexity: O(1) — no extra significant memory is used.

```python
class Solution(object):
    def findNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = 0
        for i in range(len(nums)):
            if len(str(nums[i])) % 2 == 0:
                count += 1
        return count
```

## Rubber Duck Explanation

We start with a counter set to zero.
This will track how many numbers have an even number of digits.

We loop through every number in nums.
This loop lets us inspect each value one-by-one.

For each number, we convert it to a string. `str(nums[i])`  Why?
Because counting digits in a number is much easier when it’s a string.

We check if the number of digits is even. `len(str(nums[i])) % 2 == 0` len(...) gives us the number of characters (digits).

% 2 == 0 checks if that amount is divisible by 2 — meaning it’s even.

If the number has an even number of digits, we increment our counter.
After checking all numbers, we return the total.
