## Problem: Average Salary Excluding the Minimum and Maximum

You are given an array of **unique integers** `salary`, where `salary[i]` represents the salary of the *i-th* employee.

Your task is to **compute the average salary**, excluding:
- the **minimum salary**, and
- the **maximum salary**.

The answer is considered correct if it is within **10⁻⁵** of the actual value.

---

## Examples

### Example 1
**Input:**
salary = [4000, 3000, 1000, 2000]

**Output:**

**Explanation:**
- Minimum salary = 1000
- Maximum salary = 4000
- Remaining salaries = [2000, 3000]
- Average = (2000 + 3000) / 2 = **2500**

---

### Example 2
**Input:**
salary = [1000, 2000, 3000]

**Output:**
2000.00000

**Explanation:**
- Minimum salary = 1000
- Maximum salary = 3000
- Remaining salary = [2000]
- Average = 2000 / 1 = **2000**

---

## Constraints

- `3 <= salary.length <= 100`
- `1000 <= salary[i] <= 10⁶`
- All salary values are **unique**


## Approach

1. **Sort the salary list**
   Sorting places the salaries in ascending order, making it easy to identify the minimum and maximum values.

2. **Exclude the minimum and maximum salaries**
   After sorting, the first element is the minimum and the last element is the maximum.
   Using slicing (`salary[1:-1]`), we extract only the middle salaries.

3. **Compute the average**
   - Calculate the sum of the remaining salaries.
   - Divide by the number of remaining elements to get the average.

4. **Return the result as a float**
   Converting to `float` ensures the output matches the required precision.

This approach is simple, clear, and efficient given the small input size constraints.


```python
class Solution(object):
    def average(self, salary):
        """
        :type salary: List[int]
        :rtype: float
        """
        salary.sort()
        middle = salary[1:-1]
        #avg
        return float(sum(middle)) / len(middle)
```
