# Problem: Buy Two Chocolates with Minimum Leftover

You are given an integer array `prices` representing the prices of various chocolates in a store, and an integer `money` representing your initial amount of money.

You must buy **exactly two chocolates** such that you still have some non-negative leftover money. Your goal is to **minimize the sum** of the prices of the two chocolates.

Return the amount of money you will have leftover after buying the two chocolates. If it's impossible to buy two chocolates without going into debt, return `money`.


### Example 1

**Input:** `prices = [1,2,2]`, `money = 3`
**Output:** `0`
**Explanation:** Buy chocolates priced at 1 and 2 units. Money left = 3 - 3 = 0.


### Example 2

**Input:** `prices = [3,2,3]`, `money = 3`
**Output:** `3`
**Explanation:** It's impossible to buy two chocolates without going into debt, so we return 3.


### Constraints

- `2 <= prices.length <= 50`
- `1 <= prices[i] <= 100`
- `1 <= money <= 100`


# Approach

1. **Sort the prices** of chocolates in ascending order.
   This ensures that the two cheapest chocolates are the first two elements.

2. **Calculate the total cost** of buying the two cheapest chocolates:
   ```python
   total_cost = prices[0] + prices[1]
   ```
3. Check if you can afford them:
- If total_cost <= money, subtract the total cost from your money and return the leftover
- Otherwise, you cannot buy two chocolates without going into debt, so return the original money

**Key Insight:** By always choosing the two cheapest chocolates, you minimize the sum of the prices and maximize the leftover money.

````python
class Solution(object):
    def buyChoco(self, prices, money):
        """
        :type prices: List[int]
        :type money: int
        :rtype: int
        """
        prices.sort()
        total_cost = prices[0] + prices[1]

        if total_cost <= money:
            return money - total_cost
        else:
            return money
````
