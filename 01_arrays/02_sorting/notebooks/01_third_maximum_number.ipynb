# Third Maximum Number

Given an integer array `nums`, return the **third distinct maximum number** in the array.

If the third distinct maximum does **not** exist, return the **maximum number** instead.


## Examples

### Example 1
**Input**
nums = [3,2,1]

**Output**
1

**Explanation**
- First distinct maximum: `3`
- Second distinct maximum: `2`
- Third distinct maximum: `1`


### Example 2
**Input**
nums = [1,2]

**Output**
2

**Explanation**
- First distinct maximum: `2`
- Second distinct maximum: `1`
- A third distinct maximum does not exist, so return the maximum (`2`).



### Example 3
**Input**
nums = [2,2,3,1]

**Output**
1

**Explanation**
- First distinct maximum: `3`
- Second distinct maximum: `2`
  (Both `2`s are counted as one since they are equal.)
- Third distinct maximum: `1`


## Constraints

- `1 ≤ nums.length ≤ 10⁴`
- `-2³¹ ≤ nums[i] ≤ 2³¹ - 1`



## Follow-up

Can you find an **O(n)** time complexity solution?


## Approach

This solution uses a **set** and **sorting** to find the third distinct maximum number.


### Step-by-Step

1. **Remove duplicates**
   Convert the list to a set to get only **distinct values**:
   ```python
   unique = list(set(nums))

2. Sort in descending order
3. Check the number of distinct elements
- If there are fewer than 3 distinct numbers, return the maximum:

## Complexity Analysis

Time Complexity:

- O(n) for creating the set + O(n log n) for sorting → overall O(n log n)

- Space Complexity: O(n) for the set and unique list

````python
class Solution(object):
    def thirdMax(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        unique = list(set(nums))
        unique.sort(reverse = True)

        if len(unique) < 3:
            return unique[0]
        else:
            return unique[2]
````
