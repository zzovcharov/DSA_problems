## Problem Description

There is a biker going on a road trip. The road trip consists of **n + 1 points** at different altitudes.
The biker starts the trip at **point 0** with an altitude of **0**.

You are given an integer array `gain` of length `n`, where `gain[i]` represents the **net gain in altitude** between point `i` and point `i + 1`.

Your task is to **return the highest altitude** the biker reaches during the trip.

---

## Examples

### Example 1
**Input:**
gain = [-5, 1, 5, 0, -7]


**Output:**

1


**Explanation:**
The altitudes at each point are:
[0, -5, -4, 1, 1, -6]

The highest altitude reached is **1**.



### Example 2
**Input:**
gain = [-4, -3, -2, -1, 4, 3, 2]

**Output:**

0


**Explanation:**
The altitudes at each point are:

[0, -4, -7, -9, -10, -6, -3, -1]

The highest altitude reached is **0**.



## Constraints

- `n == gain.length`
- `1 <= n <= 100`
- `-100 <= gain[i] <= 100`

---

## Hint

Track the **current altitude** while iterating through the `gain` array and update the **maximum altitude** seen so far.


### Approach

- Start at altitude **0**, since the biker begins at point 0.
- Maintain two variables:
  - `curr_alt` to track the current altitude as the biker moves along the route.
  - `max_alt` to store the highest altitude reached so far.
- Iterate through the `gain` array:
  - For each value, update the current altitude by adding the gain.
  - After each update, compare the current altitude with `max_alt` and update it if needed.
- After processing all gains, return `max_alt`, which represents the highest altitude reached during the trip.

**Time Complexity:** `O(n)`
**Space Complexity:** `O(1)`


```python
class Solution(object):
    def largestAltitude(self, gain):
        """
        :type gain: List[int]
        :rtype: int
        """
        curr_alt = 0
        max_alt = 0

        for g in gain:
            curr_alt += g
            max_alt = max(max_alt, curr_alt)
        return max_alt
```
