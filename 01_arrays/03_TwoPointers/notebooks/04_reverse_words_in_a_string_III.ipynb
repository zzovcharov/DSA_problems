# Reverse Words in a String

## Problem Description

Given a string `s`, reverse the **characters in each word** while:

- Preserving the **original word order**
- Preserving **all spaces** exactly as they appear

A word is defined as a sequence of characters separated by a single space.



## Examples

### Example 1

**Input**
```text
s = "Let's take LeetCode contest"
```

**Output**

"s'teL ekat edoCteeL tsetnoc"

### Example 2

**Input**

s = "Mr Ding"

**Output**

"rM gniD"

## Important Rules

- Each word is reversed individually

- The order of words does not change

- Spaces remain in the same positions

- No leading or trailing spaces

- Words are separated by exactly one space

## Constraints

- 1 <= s.length <= 5 × 10⁴

- s contains printable ASCII characters

- s has at least one word

- No leading or trailing spaces

- Words are separated by a single space



## Approach

1. **Split the sentence into words**
   - Use `split(" ")` to break the string into a list of words based on spaces.
   - This preserves the original word order.

2. **Reverse each word individually**
   - Iterate through the list of words.
   - Reverse each word using slicing `[::-1]`.
   - Store the reversed words in a new list.

3. **Rebuild the sentence**
   - Join the reversed words back together using `" ".join(...)`.
   - This restores the spaces exactly as in the original sentence.

4. **Return the final string**
   - The result contains reversed characters in each word, while keeping the sentence structure unchanged.



### Time & Space Complexity

- **Time Complexity:** `O(n)`
  Each character is processed once.
- **Space Complexity:** `O(n)`
  A new list is used to store the reversed words.



### Intuition (Simple Explanation)

Think of the sentence as a row of boxes (words).
You **flip each box upside down**, but **do not move the boxes themselves**.

Example:
- "Mr Ding"
- ["Mr", "Ding"]
- ["rM", "gniD"]
- "rM gniD"


Each word is reversed, but the order stays the same.


````python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        words = s.split(" ")

        reversed_words = []
        for word in words:
            reversed_words.append(word[::-1])

        return " ".join(reversed_words)
````
