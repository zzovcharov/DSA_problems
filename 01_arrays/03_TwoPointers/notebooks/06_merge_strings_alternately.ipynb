Merge the strings by adding letters in **alternating order**, starting with `word1`.

- If one string is longer than the other, append the remaining letters to the **end** of the merged string.

### Example 1

**Output:**

**Explanation:**

word2: p q r
merged: a p b q c r




### Example 2

**Input:**
word1 = "ab", word2 = "pqrs"


**Output:**
"apbqrs"


**Explanation:**
Since `word2` is longer, the extra characters `"rs"` are appended at the end.
word1: a b
word2: p q r s
merged: a p b q r s



### Example 3

**Input:**
word1 = "abcd", word2 = "pq"

**Output:**
"apbqcd"

**Explanation:**
Since `word1` is longer, the extra characters `"cd"` are appended at the end.

word1: a b c d
word2: p q
merged: a p b q c d



### Constraints

- `1 <= word1.length, word2.length <= 100`
- `word1` and `word2` consist only of lowercase English letters



## Approach

1. **Use a pointer (`i`)**
   Walk through both strings at the same time using a single index.

2. **Create a list (`merged`)**
   This helps efficiently build the final string character by character.

3. **Merge alternately**
   - While both `word1` and `word2` still have characters at index `i`:
     - Add `word1[i]`
     - Add `word2[i]`
     - Move to the next index

4. **Append remaining characters**
   - If `word1` is longer, append the remaining substring.
   - If `word2` is longer, append the remaining substring.

5. **Convert to string**
   Join the list into a single string and return it.

### Complexity
- **Time Complexity:** `O(n)` where `n` is the length of the longer string
- **Space Complexity:** `O(n)` for storing the merged result


```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        merged = []
        i = 0

        #merge
        while i < len(word1) and i < len(word2):
            merged.append(word1[i])
            merged.append(word2[i])
            i += 1

        #append the remaining part
        if i < len(word1):
            merged.append(word1[i:])
        if i < len(word2):
            merged.append(word2[i:])
        return ''.join(merged)

```
