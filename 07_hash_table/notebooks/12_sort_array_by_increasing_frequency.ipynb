## Sort Array by Increasing Frequency

Given an array of integers `nums`, sort the array in **increasing order based on the frequency** of each value.

If multiple values have the **same frequency**, sort those values in **decreasing numerical order**.

Return the sorted array.



### Example 1

**Input**

nums = [1,1,2,2,2,3]

**Output**

[3,1,1,2,2,2]


**Explanation**

- `3` appears **1 time**
- `1` appears **2 times**
- `2` appears **3 times**

Sorted by frequency: `3 → 1 → 2`



### Example 2

**Input**

nums = [2,3,1,3,2]


**Output**

[1,3,3,2,2]


**Explanation**

- `1` appears **1 time**
- `2` and `3` both appear **2 times**

Since `2` and `3` have the same frequency, they are sorted in **decreasing order** → `3` before `2`.



### Example 3

**Input**

nums = [-1,1,-6,4,5,-6,1,4,1]


**Output**

[5,-1,4,4,-6,-6,1,1,1]


### Constraints

- `1 <= nums.length <= 100`
- `-100 <= nums[i] <= 100`


## Approach — Sort Array by Increasing Frequency

We want to sort the array based on two rules:

1. **Primary rule:** Numbers with smaller frequency come first.
2. **Tie-breaker:** If two numbers have the same frequency, the larger number comes first.

**Time Complexity**

- Counting frequencies: O(n)

- Sorting: O(n log n)

- Overall: O(n log n)

**Space Complexity**

- Frequency dictionary: O(n)

```python
class Solution(object):
    def frequencySort(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        freq= {}

        for n in nums:
            freq[n] = freq.get(n, 0) + 1
        return sorted(nums, key = lambda x: (freq[x], -x))
```
